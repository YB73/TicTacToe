```python
def display_board(board_list):
    print("Board as of now:\n")
    for i in range(0,9):
        print(f'{board_list[i]}',end="")
        if i not in (2,5,8):
            print(" |",end="")
        if i in (2,5,8) and i<8 :
            print('\n--------\n')
```

```python
board_list=['1','2','3','4','5','6','7','8','9']
display_board(board_list)
```

```python
import random
def player_turn():
    return random.randint(1,2)
```

```python
def players_input():
    player_input=''
    first_turn=player_turn()
    print(f"\n\nPlayer {first_turn} will play first")
    player_input=input(f"Player {first_turn} would like to have which symbol: 'X' or '0' ")
    if player_input not in ('X','0'):
        while player_input not in ('X','0'):  
            print("Wrong/No input")
            player_input=input(f"Player {first_turn} would like to have which symbol: 'X' or '0' ")  
    return player_input
```

```python
def win():
    result=False
    for i in range(0,9):
        if i%3==0:
            if board_list[i]==board_list[i+1]==board_list[i+2] :
                result=True
        if i in (0,1,2):
            if board_list[i]==board_list[i+3]==board_list[i+6] :
                result=True
        if board_list[0]==board_list[4]==board_list[8]:
            result=True
        if board_list[2]==board_list[4]==board_list[6]:
            result=True
    return result
```

```python
def turn_values():
    input_value1=players_input()
    input_value2='X'
    if input_value1=='X':
        input_value2='0'
    while(win()!=True):    
        input1=int(input("\nEnter the index where you want to place your symbol (from 1 to 9)"))
        board_list[input1-1]=input_value1
        display_board(board_list)
        if win()== True:
            print("Player with first turn wins")
            break    
        input2=int(input("\nEnter the index where you want to place your symbol (from 1 to 9)"))
        board_list[input2-1]=input_value2
        display_board(board_list)
        if win()== True:
            print("Player with second turn wins")
            break    
    print("WIN !!")
```
